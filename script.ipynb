# [dbo].[SEL_ALUMNOS_CURSOR]

```sql
/****** Object:  StoredProcedure [dbo].[SEL_ALUMNOS_CURSOR]    Script Date: 28/9/2024 15:44:10 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
-- =============================================
-- Author:		<Author,,Name>
-- Create date: <Create Date,,>
-- Description:	<Description,,>
-- =============================================
CREATE PROCEDURE [dbo].[SEL_ALUMNOS_CURSOR] 
	
AS
BEGIN
	DECLARE cAlumnos CURSOR FOR 
		SELECT MATALU, APEALU, NOMALU
			FROM Alumnos
	DECLARE @MATALU INT
	DECLARE @APEALU VARCHAR(500)
	DECLARE @NOMALU VARCHAR(500)
	OPEN cAlumnos 
	FETCH NEXT INTO @MATALU, @APEALU, @NOMALU
	WHILE @@FETCH_STATUS = 0
		BEGIN
			SELECT @MATALU, @APEALU, @NOMALU
			FETCH NEXT INTO @MATALU, @APEALU, @NOMALU
		END
	CLOSE cAlumnos
	DEALLOCATE cAlumnos
END
GO
```
